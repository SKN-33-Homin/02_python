# for 반복문

- `for` 반복문은 iterable 객체의 요소를 하나씩 꺼내 실행한다.
- `str`, `list`, `tuple`, `dict`, `set`, `range` 등이 대표적인 iterable 객체이다.
- 반복 횟수가 정해져 있거나 컬렉션의 모든 요소를 순회할 때 자주 사용한다.

## 1. 문자열, 리스트, 튜플 순회

문자열은 문자 단위로, 리스트와 튜플은 요소 단위로 순회한다.

```python
from unittest import result

# for 객체 값을 저장할 변수 in 순회 가능한 객체

txt: str = "Hello World"
for ch in txt:
    print(ch)

lst: list[int] = [10, 20, 30]

for num in lst:
    print(num)

tpl: tuple[int, ...] = (1, 2, 3)

for num in tpl:
    print(num)
```

## 2. dict와 set 순회

`dict`를 그대로 순회하면 key가 나온다. `items()`를 사용하면 key와 value를 함께 꺼낼 수 있다.

`set`은 순서가 없는 자료형이므로 출력 순서는 실행 환경에 따라 달라질 수 있다.

```python
user_info: dict[str, object] = {

    'name' : '홍길동',
    'age' : 23,
    "hobby" : ['음악','독서','유튜브']
}

for key in user_info.keys():
    print(f'Key = {key}')
print('\n')
for key, value in user_info.items():
    print(f'Key = {key}, Value = {value}')

print('\n--- set 순회 (if를 곁들인) ---')
nums: set[int] = {1, 2, 3, 4, 5, 6}
result: list[int] = []

for num in nums:
    if num % 2 == 0:
        result.append(num)

print(result)
```

## 3. range 객체

`range(start, end, step)`은 정수 범위를 표현하는 객체이다.

- `start`: 시작값
- `end`: 끝값, 포함되지 않음
- `step`: 증가 또는 감소 간격

```python
# range 객체 : 지정된 범위 내 step만큼 증가하며 값을 하나씩 반환

rng = range(1, 5, 1)
print(rng, type(rng))

for num in rng:
    print(num, end=' ')

print('\n--- range 객체를 이용한 정수 리스트 만들기')

numbers: list[int] = list(range(1, 11, 1))
print(numbers)


print('\n--- 1부터 100까지 모든 정수의 합 ---')

n_sum = 0 # for문의 영향을 받지 않아 누적 가능

for n in range(1, 101):
    n_sum += n

print(n_sum)

print('\n--- 1부터 100까지 홀수의 합 구하기 ---')

even_sum = 0
for n in range(1, 101, 2):
    even_sum += n

print(even_sum)
```

## 4. 중첩 반복문

반복문 안에 반복문을 넣으면 행과 열처럼 2차원 구조를 만들 수 있다. 구구단과 별찍기 같은 예제에 자주 사용한다.

```python
print("--- 중첩 반복문 예시 ---")

for i in range(1,4):
    for j in range(1,3):
        print(i, j)

```

```python
dan: int = 2
for j in range(1, 10):
    print(f'{dan} * {j} = {dan * j}')
```

```python
for i in range(2, 10):
    for j in range(1, 10):
        print(f'{i} * {j} = {i * j}')
```

```python
print('--- range를 활용한 반복 횟수 설정 ---')

print("# 3행 5열")
for row in range(3):
    for col in range(5):
        print("#", end = " ")
    print("")
```

## 5. zip과 enumerate

- `zip()`은 여러 iterable을 같은 순서끼리 묶는다.
- `enumerate()`는 인덱스와 요소를 함께 꺼낸다.

```python
print('--- zip 객체 ---')
names: list[str] = ['a', 'b', 'c']
nums: list[int] = [10, 20, 30]

z = zip(names, nums)
print(z)

for name, num in z:
    print(f'{name}, {num}')

print('\n--- enumerate 객체 ---')

foods: list[str] = ['닭강정', '치킨', '짜장면', '탕수육']

for idx, food in enumerate(foods):
    print(f'{idx}번 인덱스 : {food}')
```

## 6. 반복문으로 평균 구하기

반복문을 사용하면 리스트나 딕셔너리 리스트에 들어 있는 값을 누적해서 평균을 계산할 수 있다.

```python
print('--- list에 담긴 값 평균 구하기 ---')

n_sum = 0

scores: list[int] = [20, 40, 50, 80, 100]

for score in scores:
    n_sum += score

score_avg = n_sum/len(scores)

print(score_avg)



```

```python
print('--- dict에 담긴 값 평균 구하기 ---')

students: list[dict[str, object]] = [
    {'name': '홍길동', 'score': 100},
    {'name': '신사임당', 'score': 77},
    {'name': '이순신', 'score': 85},
]

dct_sum: int = 0

# for student in students:
#     dct_sum += student.get('score',0) 리스트 순회 이용

for idx in range(len(students)):
    dct_sum += students[idx]['score'] # range와 index 이용

avg = dct_sum/len(students)
print(round(avg, 2))
```

# dict(Dictionary 사전 자료형)
- `dict`는 하나의 요소를 Key : Value 형식으로 갖는 자료형
- 사전 처럼 단어를 알면 뜻을 알 수 있듯이, Key를 통해 Value를 확인하는 구조
- list처럼 index 접근이 아닌 `Key`를 이용하여 접근

## 1. dict 생성과 key 규칙

`dict`의 핵심은 `key`이다.

- 하나의 `dict` 안에서 `key`는 중복될 수 없다.
- 같은 `key`를 다시 넣으면 마지막 값으로 덮어써진다.
- `key`는 값을 빠르게 찾기 위한 기준이므로 중간에 값이 바뀌면 안 된다.
- 그래서 `int`, `float`, `str`, `tuple` 같은 불변 자료형은 `key`로 사용할 수 있지만, `list`, `dict`, `set` 같은 가변 자료형은 `key`로 사용할 수 없다.
- Python 3.7 이상에서 `dict`는 삽입 순서를 유지한다.

```python
from pygments.styles import fruity

print("--- dict ---")
dct = {
    'a' : 10,
    'b' : 20,
    'c' : 100
}
print(dct, type(dct))

print(dct['a'], dct['b'], dct['c'])

# Key는 불변 타입만 가능
dct_2 = {
    10 : 'aaa',
    (10,20) : "bbb"
}
print(dct_2, type(dct_2))

#dct_3 = {
#    [1,2,3] = 'ccc',
#} 가변 타입 불가
```

```python
print('--- dict 조회 ---')

print(dct)

# Key를 이용한 조회
print(dct['a'])
print(dct.get('a'))

# 없는 키 조회
# print(dct['d']) KeyError
print(dct.get('d')) # 존재하지 않으면 None

# 없으면 get을 이용하여 적용
print(dct.get('d', '값 없음'))
print(dct.get('c', '값 없음'))
```

## dict 요소 추가, 수정, 삭제
- dict는 mutable 자료형
- dict[존재하지 않는 Key] = Value 넣을 시 `존재하지 않는 Key : Value`
- dict[Key] = Value 을 이용하여 Value 값 수정 가능
- dict.update()를 사용하여 Value 수정이 가능(동시 수정 가능)
- dict.pop() 또는 del 사용하여 요소 제거 가능

```python
dct['e'] = 888
print(dct)

dct['e'] = 999
print(dct)

dct.update(e = 1234)
print(dct)

#여러 Value 업데이트
dct.update({'a':111, 'b':222})
print(dct)

#요소 제거
del dct['c']
print(dct)

# pop()
temp = dct.pop('e')
print(dct)
print(temp)
```

```python
# dict 생성 함수와 주요 메서드
# - dict 생성 방법
#  1) {K:V, K:V,...}
#  2) dict(key=value,...) 함수를 이용
dct_3 = dict(name = '홍길동', age = 22)
print(dct_3)

# 주요 메서드
keys = dct_3.keys()
print(keys)

values = dct_3.values()
print(values)

tms = dct_3.items() # item == k:v 현태의 한쌍
print(tms, type(tms))
```

# dict 순회와 unpacking
`dict` 반복문(for 문)에 바로 넣으면 key만 반환
`value` 또는 `item`을 반환 받기 위해선 values, items 사용해야함

```python
dct_4 = {
    '이순신': 30,
    '홍길동': 20,
    '심사임당': 40,
    '강감찬': 60
}

print('--- dict 순회 기본 ---')

for key in dct_4:
    print(f'이름 : {key}, / 나이 : {dct_4[key]}세')

print('--- dict.keys ---')

for key in dct_4.keys():
    print(f'이름 : {key}, / 나이 : {dct_4[key]}세')

print('--- dict.values ---')

for value in dct_4.values():
    print(value)

print('--- dict.items ---')

for key, value in dct_4.items():
    print(key, value)

for item in dct_4.items():
    print(item)
```

# dict 대입, 얕은 복사, 깊은 복사

- 대입: 같은 객체를 함께 바라보도록 참조를 복사한다.
- 얕은 복사(shallow copy): 새 컨테이너 객체를 만들지만, 내부 중첩 객체는 공유할 수 있다.
- 깊은 복사(deep copy): 내부 중첩 객체까지 재귀적으로 새로 복사한다.
- 중첩 객체가 없는 단순 dict에서는 copy()가 깊은 복사처럼 보일 수 있지만, 정확히는 얕은 복사이다.

```python
sample = {
    'name':'기계식 키보드',
    'price': 30000,
    'origin':"Korea"
}

print('--- dict 대입 : 새 변수에 주소 복사(같은 주소 참조) ---')

other = sample # 얕은 복사 주소 값이 닽으며 같은 컨테이너
print('sample id :', id(sample))
print('other id :', id(other))

sample['name'] = '비싼 키보드'

print('before sample[name] :', sample['name'])
print('before other[name] :', other['name'])

print('\nafter sample[name] :', sample['name'])
print('after other[name] :', other['name'])
```

```python
print('--- dicr.copy() : 얕은 복사---')
print('얕은 복사, 주소 값이 다르며 다른 컨테이너 불변 타입의 경우 새 주소를 참조하기에 공유 불가, 가변 타입의 경우 주소가 유지 되기에 공유 가능')
other_2 = sample.copy() # 얕은 복사, 주소 값이 다르며 다른 컨테이너 불변 타입의 경우 새 주소를 참조하기에 복사 불가
                        # 가변 타입의 경우 주소가 유지 되기에 두 가지 변수 모두 변경

print('sample id :', id(sample))
print('other_2 id :', id(other_2))

sample['name'] = '싼 키보드'

print('before sample[name] :', sample['name'])
print('before other_2[name] :', other_2['name'])

print('\nafter sample[name] :', sample['name'])
print('after other_2[name] :', other_2['name'])

other_2['sds'] =10000
print('\n')
print(sample)
print(other_2)

sample['sd3'] =10000000
print('\n')
print(sample)
print(other_2)
```

```python
print('--- copy.deepcopy(dict) : 객체 복사 ---')

import copy

fruits = {
    'name' : 'apple',
    "price": 2000,
    'options':['red','200g']
}

other_3 = copy.deepcopy(fruits)

print(fruits)
print(other_3)
print("")
print('fruits id :', id(sample))
print('other_3 id :', id(other_3))
print("")
fruits['price'] = 3000
fruits['options'].append('상주')
print(fruits)
print(other_3)
print("")

```

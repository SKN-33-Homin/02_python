# 유효범위 Scope

변수의 유효범위(scope)는 변수에 접근할 수 있는 범위를 의미한다.

- 전역 변수(global variable): 함수 블록 밖에서 선언되어 여러 위치에서 접근할 수 있는 변수
- 지역 변수(local variable): 함수 내부에서 선언되어 해당 함수 안에서 주로 사용하는 변수
- Python의 `for`, `if` 블록은 Java의 중괄호 블록처럼 별도의 지역 스코프를 만들지 않는다.


## 1. 전역 변수와 지역 변수

함수 밖에서 선언된 `a`는 전역 변수이고, 함수 안에서 선언된 `b`는 지역 변수이다.
`b`는 `foo()` 함수 내부에서만 사용할 수 있다.


```python
a = 100 # Global : 어디서든 사용가능한 변수

def foo():
    print(f"foo에서 출력한 a : {a}")
    b = 200 # Local : 함수 내부에서만 호출 가능한 변수

    print(f"foo에서 출력한 b : {b}")

foo()
print('함수 밖 a :', a)
# print('함수 밖 b :', b) NameError: name 'b' is not defined
```

## 2. for, if 블록과 스코프

Python에서는 `for`, `if` 블록이 별도의 지역 스코프를 만들지 않는다.
따라서 블록 안에서 만들어진 이름을 블록 밖에서도 참조할 수 있다.


```python
a = 0

for i in range(3):
    print('for 문 안 i :', i)
    j = 100

print('for 문 밖 i :', i) # for문은 Local 생성 하지 않음
print('for 문 밖 j :', j)

if True:
    k = 200

print('if 문 밖 k :', k)
```

## 3. 이름 탐색 우선순위

같은 이름의 변수가 여러 범위에 있을 때는 더 가까운 범위의 변수가 우선된다.
함수 안에서 같은 이름으로 대입하면 전역 변수를 바꾸는 것이 아니라, 지역 변수를 새로 만든다.


```python
x = 100 #Global

def bar():
    x = 200 #Local : 가까운 범위 우선 생성 - Globar에 있어도 Local 내부에 새로 생성하여 적용
    print(x)

bar()
print(x)
```

## 4. 전역 변수와 지역 변수가 함께 사용되는 경우

아래 예제에서 `koo()` 함수 안의 `glo`는 지역 변수이고, 함수 밖의 `glo`와는 다른 이름 공간에 있다.


```python
glo = 10 # Global
bal = 5

def koo(a: int) -> int:
    glo = a + 10 # Global의 glo와 이름만 같은 Local
    print("koo 함수 내 glo : ", glo)
    b = a + glo +bal # Lpcal + Locla + Gloval
    return b

koo(7)
print(koo(7))
print("koo 함수 밖 glo : ", glo)
```

## 5. immutable, mutable 객체 전달

Python은 함수에 값을 전달할 때 객체에 대한 참조를 전달한다.
다만 immutable 객체는 내부 값을 직접 바꿀 수 없고, mutable 객체는 내부 값을 바꿀 수 있어 결과가 다르게 보인다.

- immutable: `int`, `float`, `bool`, `tuple`, `str`
- mutable: `list`, `dict`, `set`


```python
def func1(a :int):
    a = 200 # 전달 받은 a에 새 값 200을 대입(수정)
    print('fucn1 내 a', a, id(a)) # Local 내 새로운 지역변수가 생성 후 제거

a = 100
print("호출 전 a", a, id(a))

func1(a)
print("호출 후 a", a, id(a))
```

```python
def func2(b: list) -> None:
    b[0] = 5 # 전달 받은 a에 새 값 200을 대입(수정)
    print('fucn2 내 b', b, id(b)) # Local 내 새로운 지역변수가 생성 후 제거

b = [1,2,3]
print("호출 전 b", b, id(b))

func2(b)
print("호출 후 b", b, id(b))
```

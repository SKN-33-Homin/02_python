## Getter/Setter 메서드
캡슐화된 클래스에서 속성을 제어하기 위한 메서드로써 기본적으로 다음 두 유형이 있다.
1. getter 속성의 값을 읽어내기 위한 메서드 (외부에서 값 가져감)
2. setter 속성의 값을 설정하기 위한 메서드 (외부에서 값 대입받음)

```python
class my_class :

    def __init__(self, value) :
        self.__value = value

    def get_value(self) :
        return self.__value

    def set_value(self, value) :
        self.__value = value

my_cls = my_class(100)

print(my_cls.get_value())
my_cls.set_value(200)
print(my_cls.get_value())
```

## Python의 `@property`

Python에서는 Java식 `get_value()`, `set_value()` 메서드도 사용할 수 있지만, 더 Python다운 방식으로는 `@property`를 자주 사용한다.

`@property`를 사용하면 메서드로 값을 검증하면서도, 사용하는 쪽에서는 속성처럼 읽고 쓸 수 있다.


```python
class Score:
    def __init__(self, value: int) :
        self.__value = value

    # 오버로딩: 한 클래스내에 같은 이름의 메서드가 여러개 존재하도록 하는 기능
    @property #getter
    def value(self): # 위의 get_value와 같음
        return self.__value

    @value.setter #setter
    def value(self, value: int) :# set_value
        if not (100 >= value >= 0):
            raise ValueError("점수는 0~100 사이만 가능")
        self.__value = value

score = Score(80)
print(score.value)

# score.value(90): 에러
score.value = 90 # 메서드 호출이 아닌 변수에 값을 대입

print(score.value)

score.value = 120
```

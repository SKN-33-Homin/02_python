# dict(Dictionary 사전 자료형)
- `dict`는 하나의 요소를 Key : Value 형식으로 갖는 자료형
- 사전 처럼 단어를 알면 뜻을 알 수 있듯이, Key를 통해 Value를 확인하는 구조
- list처럼 index 접근이 아닌 `Key`를 이용하여 접근

## 1. dict 생성과 key 규칙

`dict`의 핵심은 `key`이다.

- 하나의 `dict` 안에서 `key`는 중복될 수 없다.
- 같은 `key`를 다시 넣으면 마지막 값으로 덮어써진다.
- `key`는 값을 빠르게 찾기 위한 기준이므로 중간에 값이 바뀌면 안 된다.
- 그래서 `int`, `float`, `str`, `tuple` 같은 불변 자료형은 `key`로 사용할 수 있지만, `list`, `dict`, `set` 같은 가변 자료형은 `key`로 사용할 수 없다.
- Python 3.7 이상에서 `dict`는 삽입 순서를 유지한다.

```python
print("--- dict ---")
dct = {
    'a' : 10,
    'b' : 20,
    'c' : 100
}
print(dct, type(dct))

print(dct['a'], dct['b'], dct['c'])

# Key는 불변 타입만 가능
dct_2 = {
    10 : 'aaa',
    (10,20) : "bbb"
}
print(dct_2, type(dct_2))

#dct_3 = {
#    [1,2,3] = 'ccc',
#} 가변 타입 불가
```

```python
print('--- dict 조회 ---')

print(dct)

# Key를 이용한 조회
print(dct['a'])
print(dct.get('a'))

# 없는 키 조회
# print(dct['d']) KeyError
print(dct.get('d')) # 존재하지 않으면 None

# 없으면 get을 이용하여 적용
print(dct.get('d', '값 없음'))
print(dct.get('c', '값 없음'))
```

## dict 요소 추가, 수정, 삭제
- dict는 mutable 자료형
- dict[존재하지 않는 Key] = Value 넣을 시 `존재하지 않는 Key : Value`
- dict[Key] = Value 을 이용하여 Value 값 수정 가능
- dict.update()를 사용하여 Value 수정이 가능(동시 수정 가능)
- dict.pop() 또는 del 사용하여 요소 제거 가능

```python
dct['e'] = 888
print(dct)

dct['e'] = 999
print(dct)

dct.update(e = 1234)
print(dct)

#여러 Value 업데이트
dct.update({'a':111, 'b':222})
print(dct)

#요소 제거
del dct['c']
print(dct)

# pop()
temp = dct.pop('e')
print(dct)
print(temp)
```

```python
# dict 생성 함수와 주요 메서드
# - dict 생성 방법
#  1) {K:V, K:V,...}
#  2) dict(key=value,...) 함수를 이용
dct_3 = dict(name = '홍길동', age = 22)
print(dct_3)

# 주요 메서드
keys = dct_3.keys()
print(keys)

values = dct_3.values()
print(values)

tms = dct_3.items() # item == k:v 현태의 한쌍
print(tms, type(tms))
```

### 응집도
- 해당 클래스와 연관된 코드만 사용(코드 자원 낭비 방지) -> 하나의 목적을 위한 코드 (높을수록 좋음)

### 결합도
- 코드간의 의존도 (낮을수록 좋음)

### SOLID 원칙
- SRP (단일 책임 원칙): 하나의 클래스는 하나의 책임만 가져야 하며, 변경 이유는 단 하나여야 합니다.
- OCP (개방-폐쇄 원칙): 확장에는 열려 있으나 변경에는 닫혀 있어야 합니다. 인터페이스나 다형성을 활용하여 기존 코드 수정 없이 기능을 추가합니다.
- LSP (리스코프 치환 원칙): 하위 타입 인스턴스로 부모 타입을 대체해도 프로그램의 정확성이 유지되어야 합니다.
- ISP (인터페이스 분리 원칙): 클라이언트가 사용하지 않는 메서드에 의존하지 않도록, 인터페이스를 필요한 만큼 작게 분리해야 합니다.
- DIP (의존 역전 원칙): 구체화된 클래스가 아닌 인터페이스(추상화)에 의존하여 상위 모듈과 하위 모듈 간의 결합도를 낮춥니다.

## Driving Application

### 1. 요구사항 명세 (클라이언트)
- 운전프로그램을 만들어 주세요.
- 운전자는 시동걸기/끄기, 악셀 또는 브레이크를 밟을 수 있습니다.
- 자동차는 엔진시작/끝, 가속/감속을 할 수 있습니다.
- 자동차는 처음에 대기상태 있어야 합니다. (시동이 꺼진 상태)
- 자동차는 운전자에 의해 시동이 걸리고, 이미 시동이 걸려있다면, 또 시동을 걸수는 없습니다.
- 운전자가 악셀을 밟으면, 자동차는 10km/h씩 가속할 수 있습니다.
- 최대속도는 200km/h 입니다.
- 운전자가 브레이크를 밟으면, 자동차는 10km/h씩 감속할 수 있습니다.
- 운전자가 시동을 끄면, 자동차는 더이상 움직일 수 없습니다. (시동이 꺼진 상태)
- 자동차가 달리는 동안에는 시동을 끌수 없습니다.

### 2. 객체 도출
- 운전자

### 3. 구현 포인트

요구사항은 단순히 속도만 올리고 내리는 것이 아니라, 현재 자동차의 상태를 확인하면서 동작을 제한해야 한다.

- 시동이 꺼져 있으면 가속/감속 불가
- 이미 시동이 걸려 있으면 다시 시동 걸기 불가
- 속도가 0보다 크면 시동 끄기 불가
- 최대 속도를 초과하지 않도록 제한

노트북에서는 `input()`으로 실행을 멈추지 않도록 샘플 메뉴 번호를 전달해 실행한다.


```python
"""
Car class 생성

1. Car 객체 생성
2-1. 처음 시작 상태는 무조건 시동 꺼짐 상태
2-2. Car 내부에 엔진 출발 옵션 - 시동이 4멈췄을 때만 가능 하며 시동이 걸려야 이후 동작 가능
3. Car 내부에 엔진 멈춤 옵션 - 속도가 0일때만 동작 가능
4. Car 내부에 엔진 출력 상승 옵션 - 최대 속도 200까지만 가능
5. Car 내부에 엔진 출력 하강 옵션 - 최소 0까지만 가능

Engine class 생성

1. 현재 엔진의 상태
2. 속도 증가
3. 속도 감소
"""
class Engine:

    def __init__(self):
        self.status = 0
        self.start = False

    def get_status(self):
        return f'현재 속도 {self.status}입니다.'

    def up_speed(self):
        if self.status >= 200:
            return f'속도를 더 이상 올릴 수 없습니다.'
        else:
            self.status += 10
            return f'현재 속도 {self.status}'

    def down_speed(self):
        if self.status == 0:
           return f'속도를 더 이상 내릴 수 없습니다.'
        else:
            self.status -= 10
            return f'현재 속도 {self.status}'

class Car:

    def __init__(self, number: str, engine: Engine):
        self.__number = number
        self.engine = Engine()

    def start_engine(self):
        if not self.engine.start:
            self.engine.start = True
            return f'운전을 시작합니다.'
        else:
            return f'시동이 켜진 상태입니다.'

    def stop_engine(self):
        if not self.engine.start:
            return f'시동이 꺼진 상태입니다.'
        elif not self.engine.status == 0:
            return f'속도를 완전히 줄여주시기 바랍니다.'
        else:
            self.engine.start = False
            return f'운전을 종료합니다.'

    def increase_speed(self):
        if not self.engine.start:
            return f'시동이 꺼진 상태입니다.'
        else:
            return self.engine.up_speed()

    def decrease_speed(self):
        if not self.engine.start:
            return f'시동이 꺼진 상태입니다.'
        else:
            return self.engine.down_speed()

    def get_car_status(self):
        return self.engine.get_status()

    def __str__(self):
        return f'현재 시동이 켜져있습니다.' if self.engine.start else f'현재 시동이 꺼진 상태입니다.'
```

```python
class Driver:

    def __init__(self, car_num: str, engine_str: Engine):
        self.__car = Car(car_num, engine_str)

    def get_car_status(self):
        return self.__car.get_car_status()

    def start_car(self):
        return self.__car.start_engine() # Car 내부에 엔진 출발 옵션

    def stop_car(self):
        return self.__car.stop_engine() # Car 내부에 엔진 멈춤 옵션

    def accelerate(self):
        return self.__car.increase_speed() # Car 내부에 엔진 출력 상승 옵션

    def brake(self):
        return self.__car.decrease_speed() # Car 내부에 엔진 출력 하강 옵션

    def __str__(self):
        return self.__car.__str__()

"""
Driver class 추가 사항
1. 현재 차 상태 확인
2. 현재 속도 확인
"""
```

```python
"""
UI class 생성

1. 시동 켜기
2. 시동 끄기
3. 속도 올리기
4. 속도 내리기
5. 현재 속도와 차 상태 확인
"""

class UI:
    def __init__(self, car_num: str, engine_str: Engine):
        self.__driver = Driver(car_num, engine_str)

    def match_number(self, number: int):
        if number == 1:
            return self.__driver.start_car()
        elif number == 2:
            return self.__driver.stop_car()
        elif number == 3:
            return self.__driver.accelerate()
        elif number == 4:
            return self.__driver.brake()
        elif number == 5:
            return f'{self.__driver.get_car_status()} : {self.__driver.__str__()}'
        else:
            return f'동작 할 수 없는 번호입니다.'
```

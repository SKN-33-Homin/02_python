# 조건문 if

- 조건문은 특정 조건에 따라 실행할 코드를 선택하는 문법이다.
- `if`는 조건이 참일 때만 실행한다.
- `if...else`는 둘 중 하나를 반드시 실행한다.
- `if...elif...else`는 여러 조건 중 하나를 선택한다.

## 1. if

`if`는 특정 조건에서만 실행되어야 하는 코드가 있을 때 사용한다.

예를 들어 시험 점수가 90점 이상일 때만 커피 쿠폰을 지급할 수 있다.

```python
from jedi.inference.compiled import value


# give_coupon 선언
# 매개변수 score: int
# 반환 None

def give_coupon(score: int) -> None:
    if score >= 100:
        print("커피 쿠폰 드셔요")


give_coupon(100) # 함수 호출 및 호출
give_coupon(90)


```

## 2. if...else

`if...else`는 조건이 참인 경우와 거짓인 경우를 모두 처리해야 할 때 사용한다.

예를 들어 점수가 60점 이상이면 합격, 그렇지 않으면 불합격으로 나눌 수 있다.

```python
def pass_or_fail(score: int) -> None:
    if score >= 60:
        print(f"{score}점으로 합격")
    else:
        print(f"{score}점으로 불합격")
    print("수고하셨습니다.")


pass_or_fail(50)
pass_or_fail(70)
```

## 3. if...elif...else

`if...elif...else`는 고려해야 하는 조건이 여러 개인 경우 사용한다.

학점 계산처럼 여러 구간 중 하나를 선택해야 할 때 적합하다.

```python
def get_grade(score: int) -> str:

    # 유효성 검사(validation)
    # - 제공 받은 값이 기능 수행에 있어 올바른가 확인

    if not (100 >= score >= 0):
        raise ValueError("잘못됨")
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade = "F"
    return grade
result = get_grade(50)
print(result)
print(get_grade(60))
print(get_grade(70))
print(get_grade(80))
print(get_grade(90))
```

## 4. 입력값 유효성 검사

사용자 입력은 항상 올바르다고 가정하면 안 된다. 점수처럼 범위가 정해진 값은 조건문으로 유효성을 먼저 확인하는 습관이 좋다.

```python
score =int(input("점수 입력"))

try: #에러가 발생 가능한 부분에 시도
    grade = get_grade(score)
    print(f"{score} -> {grade}등급이다.")
except ValueError as e: #에러 발생시 어떻게 할 지
    print(e)
```

```python
# with 절
# - with 옆에 작성된 open 구문에서 사용한 파일 연결을 with 절 종료와 함꼐 자동으로 close

with open('helloworld_2.txt', 'w', encoding='utf-8') as f:
    f.write("나는 야인이 될거야 어떤 시련이 몰아쳐도")
```

```python
with open('helloworld.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
# 한 줄씩 읽기
with open('helloworld.txt', 'r', encoding='utf-8') as f:
    for idx, f_read in enumerate(f):
        print(idx+1,f_read, end='')
```

```python
# append 모드 : 파일 이어 쓰기

with open('helloworld_2.txt', 'a', encoding='utf-8') as f:
    f.write('''
ddsddd
ssdssdssdsd
sdssdsds
sdsdssdss
''')

with open('./helloworld_2.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

## 파일 읽기

```python
# 파일 복사 하기
# 1. origin file -> ram (read)
# 2. ram -> copy file (write)

with open('helloworld.txt', 'r', encoding='utf-8') as f: # origin 읽기
    text = f.read() # with 절 변수는 전역변수로 활용가능

with open('./helloworld_3.txt', 'w', encoding='utf-8') as f: # copy에 origin 쓰기
    f.write(text)

with open('./helloworld_3.txt', 'r', encoding='utf-8') as f: # copy 읽기
    print(f.read())
```

```python
# 이미지 파일 복사(아무 파일 가능)
# r,w,a는 text용
# 그 외 -rb, wb 사용
with open('image.png', 'rb') as f:
    image_byte = f.read()

with open('image_copy.png', 'wb') as f:
    f.write(image_byte)
```

## dict-csv 읽고 쓰기

```python
# csv 파일 읽기
# csv.DictReader(f), csv.DictWriter(f, fieldnames)
import csv

with open('users.csv', 'r', encoding='utf-8') as f:

    reader = csv.DictReader(f)
    users = [user for user in reader]

users
```

```python
users
```

```python
# csv 쓰기

with open('users.csv', 'w', encoding='utf-8', newline= '') as f:
    # f : 쓸 파일
    # fieldnames : 작성 헤더(컬럼명)
    writer = csv.DictWriter(f, fieldnames=['id', 'name', 'email'])

    writer.writeheader()

    users.append({'id': 400, 'name': '이순신', 'email': 'lee123@naver.com'})
    users.append({'id': 500, 'name': '강감찬', 'email': 'kang123@naver.com'})

    # 여러 줄 쓰기
    # writer.writerows(users)

    # 한 줄씩 쓰기
    for user in users:
        writer.writerow(user)
```

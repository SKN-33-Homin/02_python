## 엑셀 파일 읽기

엑셀 파일은 단순 문자열 파일이 아니라 여러 시트, 셀, 서식 정보가 들어 있는 구조화 파일이다. 그래서 `openpyxl` 같은 전용 라이브러리로 읽는다.


````python
from openpyxl import load_workbook # 엑셀을 workbook 객체로 변경
import pypdf

read_xl = load_workbook('students.xlsx', data_only=True) # data_only : 엑셀 데이터만 가져오기 (수식 결과 값)

print(read_xl.sheetnames) # 엑셀의 sheet 읽어오기

work_sheet = read_xl['students']

print(work_sheet.title) # 시트 이름
print(work_sheet.max_row) # 행 개수
print(work_sheet.max_column) # 열 개수



````

````python
# 특정 셀 값 읽어 오기

print(work_sheet['B2'].value)
print(work_sheet['D3'].value)
print(f'4행 3열 값 : {work_sheet['D3'].value}')
print(f'4행 3열 값 : {work_sheet.cell(row=4, column=2).value}')
````

````python
# 엑셀 한 줄 씩 읽기
# iter == iterator == 반복자
# iter_rows == 행 반복자 : 반복될때마다 다음 행 반환
#values_only = True (결과 값만 tuple로)

for row in work_sheet.iter_rows(values_only=True):
    print(row)
````

## PDF 파일 읽기

- PDF는 페이지, 글자 위치, 폰트, 이미지 정보가 함께 들어 있는 문서 파일이다. 그래서 텍스트 파일처럼 바로 읽기보다 `pypdf` 같은 라이브러리로 페이지 단위 텍스트를 추출한다.

- 한글 PDF는 폰트와 인코딩 방식에 따라 텍스트 추출 결과가 달라질 수 있음

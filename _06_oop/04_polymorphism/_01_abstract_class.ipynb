# 추상클래스 Abstract Class - 콘티
- 자식클래스에서 구현해야 할 추상메서드를 가진 부모클래스 (인터페이스)
    - 추상 메서드 == 메서드 정의 부분 없음 = 미완성 메서드 == 콘티
- 추상메소드는 자식클래스에서 반드시 구현해야할 의무를 가진다.
- 추상클래스는 직접 객체화 할 수 없다. instance 생성 불가

- abc.ABC클래스를 상속하면 추상클래스를 선언할 수 있다.
    - ABC == Abstract Base Class
- @abstractmethod 데코레이터로 추상메소드를 선언할 수 있다.

```python
from abc import abstractmethod, ABC

#추상 클래스
#추상 class의 목적 : 자식 클래스들에게 공통된 규약을 제공
class Product(ABC): #ABC = 상속 : 해당 클래스가 너무 추상적 또는 class로 instance 생성)

    def __init__(self, code: str, name: str, price: int):
        self.__name = name
        self.__code = code
        self.__price = price

    @abstractmethod # 추상 메소드
    def get_tax(self):
        pass #자식 클래스에 정의를 넘김

    def get_price(self):
        return self.__price

    def __str__(self):
        return f'code: {self.__code}, name: {self.__name}, price: {self.__price}'
```

```python
# 추상 클래스로 instance 생성하기
# prod = Product('1111', '노트북', 10000)
# print(prod) error 생김
# TypeError: Can't instantiate abstract class Product without an implementation for abstract method 'get_tax'
```

```python
# 추상 클래스를 상속 받은 자식 클래스
class TV(Product):
    def __init__(self, code: str, name: str, price: int, size: int):
        super().__init__(code, name, price)
        self._size = size

    # 상속 받은 추상 메서드를 재작성 == overriding
    def get_tax(self):
        return self.get_price() * 0.1

    #str overriding
    def __str__(self):
        return f'{super().__str__()}, size: {self._size}'
```

```python
tv = TV('삼성', '테레비', 10000, 85)
print(tv)
print(tv.get_tax())

tv = TV('LG', 'LG테레비', 8670000000, 100)
print(tv)
print(tv.get_tax())
```

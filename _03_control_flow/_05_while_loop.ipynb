# while 반복문

- `while` 반복문은 조건식이 `True`로 평가되는 동안 코드를 반복 실행한다.
- 조건식이 `False`가 되면 반복을 중지한다.
- 반복 횟수가 미리 정해져 있지 않고, 특정 조건이 만족될 때까지 반복해야 할 때 자주 사용한다.

## 1. 기본 while 구조

`while`은 보통 초기값, 조건식, 증감식을 함께 사용한다.

```python
i = 1
while i <= 3:
    print(i)
    i += 1
```

````python
from webbrowser import register_standard_browsers

# 10 부터 1 출력
i = 0
j = 10
while i < 10:
    print(j - i)
    i += 1

````

````python
j = 10
while j > 0:
    print(j)
    j -= 1
````

## 2. break로 반복 중지하기

`while True`는 조건식이 항상 참인 반복문이다. 이 구조를 사용할 때는 반드시 반복을 빠져나올 수 있는 `break` 조건을 만들어야 한다.
- break : 가까운 반복문 종료

````python
i = 1
while True: # True를 이용하여 무한 반복시키기
    print(i)
    i += 1
    if i > 10:
        break
````

````python
# continue : 현재 반복을 무시하고 다음으로 넘어감
for x in range(3):
    print("앞",x)

    if x == 1:
         print("if", x)
         print("continue 등장")
         continue

    print("뒤", x)
````

## 3. 메뉴 주문 예제

````python
menu = """
===============
구내 식당 메뉴
===============
1. 제육 볶음
2, 김치찌개
3. 된장찌개
4. 김밥
0. 종료(break)
===============
"""

orders: list[str] = []

print(menu)


while True:

    select = int(input("메뉴 번호를 선택하세요."))
    print("선택 번호 :",select)
    match select:
        case 1:
            print("제육 볶음을 선택했습니다.")
            orders.append("제육 볶음")
        case 2:
            print("김치찌개를 선택했습니다.")
            orders.append("김치찌개")
        case 3:
            print("된장찌개를 선택했습니다.")
            orders.append("된장찌개")
        case 4:
            print("김밥을 선택했습니다.")
            orders.append("김밥")
        case 0:
            print("주문을 종료합니다.")
            print(f"{orders} 메뉴가 선택 되었습니다.")
            break
        case _:
            print("다시 선택해주시기 바랍니다.")
````

# 형변환 (Casting)

- 형변환은 값을 다른 자료형으로 바꾸는 작업이다.
- Python에서는 상황에 따라 자동으로 바뀌는 **암묵적 형변환**과, 개발자가 직접 함수로 바꾸는 **명시적 형변환**이 있다.
- 조건문에서는 값 자체가 `True` 또는 `False`처럼 평가되기도 한다.

## 1. 암묵적 형변환

암묵적 형변환은 Python이 연산 과정에서 자동으로 자료형을 맞춰 주는 경우이다.

- `int + float`는 더 넓은 표현 범위를 가진 `float` 결과가 된다.
- `bool`은 숫자 연산에서 `True == 1`, `False == 0`처럼 처리된다.

```python
print('1 + 2.3 =', 1 + 2.3) # int(1) + float(2.3) - > float(1.0) + float(2.3) = float(3.3)
print('True + 2 =', True + 2) # True + 2 = 1 + 2 = 3
bool(1)
```

## 2. 명시적 형변환

명시적 형변환은 개발자가 `str()`, `int()`, `float()` 같은 함수를 사용해 직접 자료형을 바꾸는 방식이다.

- 문자열과 숫자는 바로 더할 수 없으므로 숫자를 문자열로 바꿔야 한다.
- `int()`로 실수를 정수로 바꾸면 소수점 아래는 버려진다.
- 숫자로 된 문자열은 `float()` 또는 `int()`로 변환해 계산에 사용할 수 있다.

```python
print('int(str(100)) + 1 =', int(str(100)) + 1)
print('"안녕" + 123 =') # 불가능 str + int
print('"안녕" + 123 =', '안녕' + str(123))

height = 173.9
print('int(height) =', int(height)) # float(173.9) -> int(173.9) -> 173
print('float(123) =', float(123)) # int(123) -> float(123) -> 123.0

print('--- 문자열 parsing ---') # 문자열을 해석하여 타입, 형태를 변경

str_number = '123456789'
print(str_number)
print(type(str_number))
print('int(str_number) =', int(str_number))
print('type(int(str_number)) =', type(int(str_number)))

print('float(str_number) =', float(str_number))

# print(int(('파싱')) parsing error
```

## 3. 논리형으로의 형변환

조건식에서는 값이 `bool`로 변환되어 평가된다.

- **truthy**: 조건식에서 `True`처럼 평가되는 값
- **falsy**: 조건식에서 `False`처럼 평가되는 값
- 일반적으로 값이 있으면 `True`, 값이 비어 있거나 `0`이면 `False`로 평가된다.
- `''`는 빈 문자열이라 `False`이고, `' '`는 공백 문자 1개가 들어 있으므로 `True`이다.

```python
print('--- bool 형변환 : True ---') # 값이 0이 아닌 값이 하나 이상 존재시 True

print(bool(1)) # 1 형변환시 True
print(bool(100))
print(bool(-100))
print(bool('abc'))
print(bool(" "))
print(bool(['sd',12]))
print(bool({2:3}))


```

```python
print('--- bool 형변환 : True ---') # 값이 0이거나 값이 없을 시 True

print(bool(0)) # 0 형변환시 True
print(bool())
print(bool(''))
print(bool([]))
print(bool({}))

```

## 4. 조건문에서의 활용

리스트, 튜플, 딕셔너리 같은 컬렉션은 요소가 있으면 `True`, 비어 있으면 `False`로 평가된다. 그래서 `len(lst) > 0`처럼 직접 길이를 비교하지 않아도 조건문에서 간단히 사용할 수 있다.

```python
lst = [1,2] # truthy == True

if lst:
    print('lst 값이 존재')
else:
    print('lst 값이 없음')
```

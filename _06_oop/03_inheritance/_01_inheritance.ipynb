# 상속 Inheritance
- 부모클래스 속성/메서드를 자식클래스에서 선언없이 물려받아 사용하는 것
- 중복된 속성/메서드를 재작성할 필요가 없다.
- 자식클래스를 부모클래스로 묶어 관리할 수 있다.
- 파이썬 다중상속 지원

```python
from IPython.lib.pretty import pretty
from six import Module_six_moves_urllib_robotparser


class Person:
    def __init__(self, name: str, phone: str) -> None:
        self.__name = name
        self.__phone = phone

    def hello(self):
        print(f'안녕하세요, 저는 {self.__name}입니다. 제 연락처는 {self.__phone}이에요.')

    def get_name(self):
        return self.__name

    def get_phone(self):
        return self.__phone

    # __str__(): print(객체)를 했을 때 어떤 문자열을 보여줄지 정하는 메서드, 주소값 대신 return 값을 보여준다.
    def __str__(self):
        return f'Person(name={self.__name}, phone={self.__phone})'

p_1 = Person('홍길동','12341234')

print(p_1)
```

```python
# Person 상속 받는 Employee 생성
class Employee(Person): # Person 상속
    def __init__(self, name: str, phone: str, company: str) -> None:
        super().__init__(name, phone) # super().__init__ : 부모 생성자의 __init__을 물려 받는다.
        self.__company = company

    # method overriding
    # 부모로 부터 받은 hello 메서드를 재작성
    def hello(self):
        super().hello()
        print(f'저는 {self.__company}에서 근무 중 이에요.')

    def __str__(self):
        return f'Employee(name={self.get_name()}, phone={self.get_phone()}, company={self.__company})'

em_1 = Employee("홍길", '123444', 'SKN')
print(em_1.get_phone())
print(em_1.get_name())

em_1.hello()

print(em_1)
```

```python
class Developer(Person):

    def __init__(self, name: str, phone: str, *langs: str) -> None: # *lang(*args) : 3번쨰 이후 요소들 모두 튜플로 묶기
        super().__init__(name, phone)
        self.__langs = list(langs)

    # 수정 가능한 기능, 언어 추가 기능
        # langs에 언어를 가져 오는데 없으면 추가 있으면 ('이미 추가 되어 있습니다.')
    def add_langs(self, new_lang: str):
        if new_lang in self.__langs:
            print(new_lang, ': 이미 존재 합니다.')
        else:
            self.__langs.append(new_lang)
            print(self.__langs)

    # langs의 언어 제거 기능
        # langs에 언어를 제거 하는데 없으면 ('이미 제거 되어 있습니다.')
    def minus_langs(self, old_lang: str):
        if not old_lang in self.__langs:
            print(old_lang, ': 이미 제거 되어 있습니다.')
        else:
            self.__langs.remove(old_lang)
            print(self.__langs)

    # hello overring
        # 부머 hello 출력 후 ('사용 가능한 언어 목록 :', self.__langs)
    def hello(self):
        super().hello()
        print('사용 가능한 언어 목록 :', self.__langs)

    # __str__() overring
        # f'Developer (name={self.get_name()}, phone={self.get_phone()}, langs={self.__langs})'
    def __str__(self):
        return f'Developer (name={self.get_name()}, phone={self.get_phone()}, langs={self.__langs})'

dev = Developer("홍길", '01022224444', 'KR', 'EN', 'JP', 'CH')
print(dev)

dev.add_langs("EN_1")

dev.add_langs("EN")

dev.minus_langs("KR")

dev.minus_langs("KR")

dev.hello()

print(dev)
```

## isinstance(), issubclass(), MRO

상속 관계에서는 객체가 어떤 클래스의 인스턴스인지, 클래스가 어떤 부모를 상속하는지 확인할 수 있다.

- `isinstance(객체, 클래스)`: 객체가 해당 클래스 또는 하위 클래스의 인스턴스인지 확인
- `issubclass(자식클래스, 부모클래스)`: 클래스 사이의 상속 관계 확인
- MRO(Method Resolution Order): 메서드를 찾는 순서


```python
print('isinstance(dev, Developer), ', isinstance(dev, Developer))
print('isinstance(dev, Person), ', isinstance(dev, Person))
print('isinstance(dev, Employee), ', isinstance(dev, Employee))
print('--------------------------------------------------------------------------------')
print('issubclass(Developer, Developer), ', issubclass(Developer, Developer))
print('issubclass(Developer, Person), ', issubclass(Developer, Person))
print('issubclass(Developer, Employee), ', issubclass(Developer, Employee))
print('--------------------------------------------------------------------------------')
#object 클래스 == 모든 클래스의 부모
# class Test: -> class Test(object) : 상속이 없을 경우 object 상속이 자동 추가

print('MRO(Developer, Person, Employee)', Developer.mro())
print('MRO(Developer, Person, Employee)', Person.mro())
print('MRO(Developer, Person, Employee)', Employee.mro())
```

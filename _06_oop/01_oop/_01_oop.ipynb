
# OOP
Object-Oriented Programming 객체지향 프로그래밍
객체(메모리조각)를 중심으로 프로그램을 작성하는 방법론

- 변수(속성), 함수(메서드)를 묶어서 하나의 단위로써 프로그래밍
- 클래스를 작성후에 이를 바탕으로 n개의 객체(instance)를 생성할 수 있다.
---
### 객체(Object)
- 사물, 동물, 식물, 무형의 것
- 인식 가능한 것
- 속성(나이, 주소 등)
- 기능(듣기, 말하기, 동작 등)
- 속성과 기능으로 이루어짐(

### 현실의 객체(Object)
- 객체 간의 상호 작용에 인한 이벤트가 발생

### 객체 지향 프로그래밍(OOP)
- 현실 객체 개념을 프로그래밍에 대입
- 현실 객체(Object) -> 속성과 기능을 컴퓨터에서 정의 -> 속성과 기능이 정의된 문서 == Class
- 클래스를 보고 컴퓨터 메모리에 구현 == 컴퓨터 내 객체(instance)
- instance 끼리 상호작용을 하기 위한 메세지 교환
- instance 끼리 호출 할 경우 결과 출력, 반환 등의 사건이 발샐 == OOP



```python
# class 만들기 == 컴퓨터 속 객체를 만들기 위해 속성과 기능을 정의한 문서

class Person:

    nationality = 'KR'

    # 생성자 함수
    # 해당 클래스를 이용해 생성된 객체의 속성과 기능을 초기화
    def __init__(self, name, age):
        # self 객체의 속성 초기화
        self.name = name
        self.age = age

    # 기능 역할 함수 정의
    def greeting(self):
        print(f"안녕하세요. 저는 {self.name}이구요, {self.age}살 입니다.")
```

## 생성자 `__init__`

생성자는 객체가 생성될 때 자동으로 실행되어, 객체가 처음 가져야 할 속성을 초기화하는 메서드이다.

Python에서는 생성자 역할을 하는 메서드를 `__init__`이라는 이름으로 정의한다.
아래 코드에서 `Person('홍길동', 33)`을 실행하면 `__init__(self, name, age)`가 자동으로 호출되고, 전달한 값이 `self.name`, `self.age`에 저장된다.

- `self`: 생성된 인스턴스 자신
- `name`, `age`: 객체를 만들 때 외부에서 전달받는 값
- `self.name = name`: 전달받은 `name` 값을 현재 객체의 속성으로 저장

즉, 생성자는 객체가 만들어지는 순간 필요한 초기 데이터를 세팅하는 초기화 담당 메서드이다.


```python
# class 이용하여 객체(instance) 생성

hong = Person('홍길동', '20') # Person class의 생성자를 이영해서 instance 생성
# hong 변수에는 실제 객체가 아닌 주소 값이 저장 되어 참조
print(hong, type(hong))
# <class '__main__.Person'> : 클래스 명이 데이터 타입
# class는 사용자 정의 데이터 타입
# 한번 만든 class는 재사용 가능
print(hong.name, hong.age)
print(hong.nationality)

lee = Person("이순신", 33)

print(lee, type(lee))
lee.greeting()
```

## self 이해하기

`self`는 메서드를 호출한 인스턴스 자신을 가리키는 매개변수이다.

아래 두 호출은 같은 의미로 볼 수 있다.

- `hong.greeting()`
- `Person.greeting(hong)`

일반적으로는 첫 번째 방식처럼 인스턴스를 통해 메서드를 호출한다.


```python
hong.greeting() # hong 변수에 저장된 주소로 찾아가 Person class에서 greeting 호출
Person.greeting(hong)

print(Person)
Person.greeting(lee)# Person 타입의 객체의 주소 전달이 필요
```

## 속성(Field)/메서드(Method)

* 클래스 속성: 인스턴스간에 공유될 수 있는 속성
* 클래스 메서드: 인스턴스간에 공유될 수 있는 메서드
---
* 인스턴스 속성: 인스턴스마다 달라지는 속성(self)
* 인스턴스 메서드: 인스턴스속성에 대해서 처리하는 메서드

### 힘수(Function)과 메서드(Method)의 차이점
- 함수  : 독립적으로 존재가 가능하며 함수명으로 호출 후 기능을 수행

- 메서드 : 객체(또는 class) 내에 정의 된 함수로 (class.method 또는 주소를 저장한 참조 변수)로 호출


```python
class SKNstudent:
    course_name: str = "Family AI Camp" # class 속성, 공유, 각 객체에서 개별로 변경 가능

    def __init__(self, nth, name):
        self.name: str = name #instance 속성, 독립적으로 생성되며 각각 수정 가능
        self.nth: int = nth #instance 속성, 독립적으로 생성되며 각각 수정 가능

# SKNstudent 객체(instance 생성)
print('--- instance 변경 전 ---')
kim = SKNstudent(33, '김유진')
choi = SKNstudent(33, '최경돈')

print(kim.nth, kim.name, kim.course_name)
print(choi.nth, choi.name, choi.course_name)

print('--- instance 변경 후 ---')
kim = SKNstudent(33, '김유진')
choi = SKNstudent(33, '최경돈')
kim.nth = 34

print(kim.nth, kim.name, kim.course_name)
print(choi.nth, choi.name, choi.course_name)

print('--- class명을 이용하여 class 속성 변경 후 ---')
kim = SKNstudent(33, '김유진')
choi = SKNstudent(33, '최경돈')
SKNstudent.course_name = "AI Camp"

print(kim.nth, kim.name, kim.course_name)
print(choi.nth, choi.name, choi.course_name)

print('--- 참조 변수를 이용하여 class 속성 변경 후 ---')
kim = SKNstudent(33, '김유진')
choi = SKNstudent(33, '최경돈')
choi.course_name = "Famliy AI Camp"

print(kim.nth, kim.name, kim.course_name)
print(choi.nth, choi.name, choi.course_name)

```

```python
class SKNstudent:
    course_name: str = "Family AI Camp" # class 속성, 공유, 각 객체에서 개별로 변경 가능

    def __init__(self, nth, name):
        self.name: str = name #instance 속성, 독립적으로 생성되며 각각 수정 가능
        self.nth: int = nth #instance 속성, 독립적으로 생성되며 각각 수정 가능

    def greeting(self): # instance 메서드: 첫번쨰 매개변수로 self를 갖는 메서드, 해당 instance의 속성을 다룰 수 있음
        print(f"안녕하세요. 저는 {self.nth}기 {self.name} 입니다.")

kim = SKNstudent(33, '김유진')
choi = SKNstudent(33, '최경돈')

kim.greeting()
choi.greeting()
```

```python
class SKNstudent:
    course_name: str = "Family AI Camp" # class 속성, 공유, 각 객체에서 개별로 변경 가능

    def __init__(self, nth, name):
        self.name: str = name #instance 속성, 독립적으로 생성되며 각각 수정 가능
        self.nth: int = nth #instance 속성, 독립적으로 생성되며 각각 수정 가능

    def greeting(self): # instance 메서드: 첫번쨰 매개변수로 self를 갖는 메서드, 해당 instance의 속성을 다룰 수 있음
        print(f"안녕하세요. 저는 {self.nth}기 {self.name} 입니다.")

    # class 메서드 생성: 첫번째 매개 변수로 cls를 받는 메서드(cls : 해당 클래스 - SKNstudent)
    @classmethod # 데코레이터: 명시적 힌트
    def goodmornig(cls):
        print(f"{cls.course_name} 여러분 좋은 아침입니다.")


kim = SKNstudent(33, '김유진')
choi = SKNstudent(33, '최경돈')

# class 메서드 확인
kim.goodmornig() # instance.classmethod
choi.goodmornig()
SKNstudent.goodmornig() # class.classmethod (classmethod이므로 class 명이 나오는 걸 권장)
```

```python
class SKNstudent:
    course_name: str = "Family AI Camp" # class 속성, 공유, 각 객체에서 개별로 변경 가능

    def __init__(self, nth, name):
        self.name: str = name #instance 속성, 독립적으로 생성되며 각각 수정 가능
        self.nth: int = nth #instance 속성, 독립적으로 생성되며 각각 수정 가능

    def greeting(self): # instance 메서드: 첫번쨰 매개변수로 self를 갖는 메서드, 해당 instance의 속성을 다룰 수 있음
        print(f"안녕하세요. 저는 {self.nth}기 {self.name} 입니다.")

    # class 메서드 생성: 첫번째 매개 변수로 cls를 받는 메서드(cls : 해당 클래스 - SKNstudent)
    @classmethod # 데코레이터: 명시적 힌트
    def goodmornig(cls):
        print(f"{cls.course_name} 여러분 좋은 아침입니다.")

    #static 메서드(정적 메서드) : 참조 변수가 들어가지 않으며 instance 상태, class 상태를 직접 다루지 않는 독립적인 기능에 사용

    @staticmethod
    def course_info(campus):
        if campus == '서초':
            print(f'서초점 주소 : 서울시 서초구')
        elif campus == '동작' :
            print(f'동작점 주소 : 서울시 동작구')
        else :
            print("잘못 입력됨")

kim = SKNstudent(33, '김유진')
choi = SKNstudent(33, '최경돈')

# static 메서드 확인
kim.course_info('동작')
choi.course_info('서초')
SKNstudent.course_info('서') # 클래스 자체 입력을 권장
```

## 클래스 속성과 인스턴스 속성 주의점

- 클래스 속성은 모든 인스턴스가 공유하는 값이다.
- 반면 인스턴스에 같은 이름의 속성을 새로 대입하면, 그 인스턴스만의 속성이 생겨 클래스 속성을 가리게 된다.


```python
class Student :
    school_name = "서일초"

    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

park = Student(name="박철수", grade='3')
kim = Student('김영희', '4')


print(park.school_name)
print(kim.school_name)
print(Student.school_name)

# class.class속성 = 값 -> 공유되는 class 값 수정
Student.school_name = '역삼초'
print()
print(park.school_name)
print(kim.school_name)
print(Student.school_name)

# instance.class속성 = 값 -> 해당 객체(instance)만 class 속성 변경

park.school_name = '서이초'
print(park.__dict__)
print(kim.__dict__)
print()
print(park.school_name)
print(kim.school_name)
print(Student.school_name)
```

  - 코드 재사용성이 좋다.
3. 다형성
  - 상속에 기반한 기술. 여러 자식클래스를 부모타입으로 제어하는 것.
4. 추상화
  - 공통된, 보편적인 속성/메소드를 추출해 내는 것. 또는 이를 통해 제어
하는 것.

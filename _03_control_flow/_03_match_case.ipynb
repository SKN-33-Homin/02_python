# match...case

- `match...case`는 값의 패턴에 따라 분기 처리하는 제어문이다.
- 다른 언어의 `switch...case`와 비슷하지만, Python의 `match...case`는 값뿐 아니라 구조 패턴도 검사할 수 있다.
- Python 3.10 이상에서 사용할 수 있다.

## 1. 기본 문법

```python
match 표현식:
    case 값1:
        실행문
    case 값2:
        실행문
    case _:
        기본 실행문
```

`case _`는 앞의 어떤 조건에도 맞지 않을 때 실행되는 기본 분기이다.

````python
#조건식 : bool typr
#표현식 : 값이 일치

def print_drink_price(num: str) -> None:
    match num:
        case '1':
            print('콜라 1000')
        case '2':
            print("사이다 900")
        case '3':
            print('몬스터 2000')
        case _:
            print("잘못 입력")
    print("이용 감사요")

print_drink_price(input("이용해"))
````

## 2. 계산기 분기

연산자처럼 정해진 값 중 하나를 선택하는 경우 `match...case`를 사용하면 분기 구조가 깔끔해진다.

````python
def Calculate(a: int,b: int,op: str) -> None:
    match op:
        case '+':
            print(a + b)
        case '-':
            print(a - b)
        case'*':
            print(a * b)
        case '/':
            print(a / b)
        case '//':
            print(a // b)
        case '%':
            print(a % b)
        case '**':
            print(a ** b)
        case _:
            print("다시 입력")

for op in ['+', '-','*','/','//','%','**','@']:
    Calculate(7, 3, op)
````

## 3. guard 조건

`case 값 if 조건` 형태를 사용하면 단순 값 비교보다 더 세밀한 조건을 줄 수 있다.

````python
def classify_number(n: int) -> None:
    match n:
        case n if n > 0:
            print("양의 정수")
        case n if n < 0:
            print("음의 정수")
        case _:
            print("0입니다.")

for n in [10, -10, 0]:
    classify_number(n)

````

## 4. tuple 패턴 검사

`match...case`는 값 하나뿐 아니라 튜플 구조도 검사할 수 있다. `_`는 해당 자리에 어떤 값이 와도 된다는 뜻이다.

````python
def inspect_tuple (value: tuple[int, int]) -> None:
    match value:
        case (1, 2):
            print('1,2')
        case (1, k) if k < 0:
            print(f'1,음수')
        case (1, _) :
            print('1, 음수 제외')
        case _:
            print('다름')

for i in ((1,2), (1,-1), (1,3),(2,7)):
    inspect_tuple(i)
````

## 5. dict 패턴 검사

딕셔너리도 특정 key와 value 패턴을 기준으로 분기할 수 있다.

````python
applicants = [
    {'type': 'dev', 'career': 15},
    {'type': 'dev', 'career': 5},
    {'type': 'devops', 'career': 0},
    {'type': 'hr'},
]

def inspect_applicant(applicant: dict[str, object]) -> None:
    match applicant:
        case {'type': 'dev', 'career': career} if career >= 10:
            print('중견 개발자 지원입니다.')
        case {'type': 'dev'}:
            print('개발자 지원입니다.')
        case {'type': 'devops'}:
            print('운영/개발자 지원입니다.')
        case _:
            print('기타 지원입니다.')

for i in applicants:
    inspect_applicant(i)
````

## 6. match...case로 학점 계산하기

`if...elif...else`로 작성했던 학점 계산도 `match...case`와 guard 조건으로 표현할 수 있다.

````python
def get_grade(score: int) -> str:
    if not (100 >= score >= 0):
        raise ValueError("잘못됨")
    # 코드 작성
    match score:
        case num if num >= 90:
            grade = 'A'
        case num if num >= 80:
            grade = 'B'
        case num if num >= 70:
            grade = 'C'
        case num if num >= 60:
            grade = 'D'
        case _:
            grade = 'F'
    return grade

for score in (50,40,60,70,80,90,100, 110):

    try: # Error가 발생할 수 있는 구문을 작성해서 시도
        grade = get_grade(score)
        print(f'{score}점 -> {grade}등급')

    except ValueError as error: # try에서 ValueError 발생 시 수행
        print(error)
````

# Comprehension 내포

- Comprehension은 한 줄짜리 반복문으로 새로운 컬렉션을 만드는 문법이다.
- 대표적으로 list comprehension과 dict comprehension이 있다.
- 반복문으로 값을 하나씩 추가하는 패턴을 더 짧고 읽기 좋게 표현할 수 있다.

- list comprehension
```
# list 생성
[표현식 for 변수 in 반복가능한객체]

# 조건 포함
[표현식 for 변수 in 반복가능한객체 if 조건식]
```

- dict comprehension
```
# dict 생성
{key표현식: value표현식 for 변수 in 반복가능한객체}

# 조건 포함
{key표현식: value표현식 for 변수 in 반복가능한객체 if 조건식}
```



## 1. for 반복문과 list comprehension 비교

먼저 일반 `for` 반복문으로 리스트를 만든 뒤, 같은 결과를 list comprehension으로 바꿔 본다.

````python
from os import name


def make_numbers_with_for() -> list[int]:
    numbers: list[int] = []
    for n in range(1,6):
        numbers.append(n)

    return numbers

lst = make_numbers_with_for()
print(lst)

# list comprehension

def make_numbers_with_comprehension() -> list[int]:
    return [n*2 for n in range(1,6)]

print(make_numbers_with_comprehension())
````

## 2. 조건식이 있는 list comprehension

`if` 조건을 붙이면 원하는 요소만 필터링해서 리스트를 만들 수 있다.

````python
def make_number3() -> list[int]:
    return [i for i in range(1,101) if i % 3 == 0]

print('3의 배수 :', make_number3())
````

````python
# 전달 된 list에서 정수 타입만 필터링 해서 반환
lst = ['d', '1', '일', 1, 2.3, 2, {"k":"v"}, True]

def filter_int_values(values: list[object]) -> list[int]:
    return [v for v in values if isinstance(v, int)] #isinstance(value, type) -> bool : 타입이 일치하는가 확인 (value == type)

print(filter_int_values(lst)) # True는 1로 저장 되기에 반환
````

## 3. 중첩 반복문 list comprehension

반복문이 중첩된 구조도 comprehension으로 표현할 수 있다. 다만 너무 복잡해지면 일반 반복문이 더 읽기 쉽다.

````python
# 중첩 버전
def make_positions() -> list[tuple[int,int]]:
    positions: list[tuple[int,int]] = []

    for row in range(3): # 0~2 행
        for col in range(5): # 0~4열
            positions.append((col, row))
    return positions

print(make_positions())
````

````python
# 중첩 comprehension 버전
def make_positions_with_comprehension() -> list[tuple[int,int]]:
    return [(col, row) for row in range(3) for col in range(5)]

print(make_positions_with_comprehension())
````

## 4. dict comprehension

`dict`도 comprehension으로 만들 수 있다.

```python
{key식: value식 for 요소 in iterable}
```

````python
def make_square_map() -> dict[int, int]:
    return {k : k**2 for k in range(1,6)}

print(make_square_map())
````

````python
# 문자열 리스트를 전달하여 {"문자열" : 문자열 길이} 형태로 반환
def make_word_length(words: list[str]) -> dict[str, int]:
    # comprehension 이용
    return {w : len(w) for w in words if len(w) > 3}
    # 일반적인 방법
    # word_dict: dict[str, int] = {}
    # for word in words:
        #word_dict[word] = len[word]
    # return word_dict

words: list[str] = ['hello', '배고파', '반복문은 왜이렇게 어려운가?']
print(make_word_length(words))
````

## 5. list와 dict 변환

`zip()`과 `dict()`를 함께 사용하면 두 리스트를 하나의 딕셔너리로 묶을 수 있다. 반대로 `dict.items()`와 unpacking을 사용하면 딕셔너리에서 key와 value 묶음을 분리할 수 있다.

````python
# 집 말고 zip
def make_age_map() -> dict[str, int]:
    names: list[str] = ["홍길동", '심사임당', '이순신']
    ages: list[int] = [20, 40, 30]

    # zip(names, ages) == ['홍길동', 20, '심사임당', 40, '이순신', 30]
    return dict(zip(names, ages))

print(make_age_map())
````

````python
# unpacking

def split_age_name() -> tuple[tuple[str,...], tuple[int,...]]: #... : 가변자로 최소 0에서 최대 n개
    info: dict[str, int] = {'홍길동': 20, '심사임당': 40, '이순신': 30}
    # print(info.items()) : dict_items([('홍길동', 20), ('심사임당', 40), ('이순신', 30)])
    # print(*info.items()) : ('홍길동', 20) ('심사임당', 40) ('이순신', 30), * : unpacking

    names, ages = zip(*info.items())
    # print(names, ages) ('홍길동', '심사임당', '이순신') (20, 40, 30)

    return names, ages

print(split_age_name())
````
